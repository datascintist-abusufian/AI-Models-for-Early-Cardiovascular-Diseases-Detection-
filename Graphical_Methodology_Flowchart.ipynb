```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(12, 8))

# Define the positions and sizes of the boxes
boxes = {
    "Data Collection": (0.05, 0.8, 0.2, 0.1),
    "Data Preprocessing": (0.3, 0.8, 0.2, 0.1),
    "Feature Selection & Engineering": (0.55, 0.8, 0.3, 0.1),
    "Model Development": (0.05, 0.6, 0.25, 0.1),
    "Class Balancing (SMOTE)": (0.35, 0.6, 0.3, 0.1),
    "Model Training": (0.7, 0.6, 0.25, 0.1),
    "Model Validation": (0.05, 0.4, 0.25, 0.1),
    "Bias Mitigation": (0.35, 0.4, 0.3, 0.1),
    "User Trust Assessment": (0.7, 0.4, 0.25, 0.1),
    "Integration into Web App": (0.05, 0.2, 0.3, 0.1),
    "Clinical Validation": (0.4, 0.2, 0.25, 0.1),
    "Deployment": (0.7, 0.2, 0.25, 0.1)
}

# Draw the boxes with labels
for label, (x, y, w, h) in boxes.items():
    ax.add_patch(patches.FancyBboxPatch((x, y), w, h, boxstyle="round,pad=0.05", edgecolor="black", facecolor="skyblue"))
    ax.text(x + w / 2, y + h / 2, label, ha="center", va="center", fontsize=10)

# Define the arrows connecting the boxes
arrows = [
    ("Data Collection", "Data Preprocessing"),
    ("Data Preprocessing", "Feature Selection & Engineering"),
    ("Feature Selection & Engineering", "Model Development"),
    ("Feature Selection & Engineering", "Class Balancing (SMOTE)"),
    ("Feature Selection & Engineering", "Model Training"),
    ("Model Development", "Model Validation"),
    ("Class Balancing (SMOTE)", "Model Validation"),
    ("Model Training", "Model Validation"),
    ("Model Validation", "Bias Mitigation"),
    ("Bias Mitigation", "User Trust Assessment"),
    ("User Trust Assessment", "Integration into Web App"),
    ("Integration into Web App", "Clinical Validation"),
    ("Clinical Validation", "Deployment")
]

# Draw the arrows
for start, end in arrows:
    sx, sy = boxes[start][0] + boxes[start][2] / 2, boxes[start][1]
    ex, ey = boxes[end][0] + boxes[end][2] / 2, boxes[end][1] + boxes[end][3]
    ax.annotate("", xy=(ex, ey), xytext=(sx, sy),
                arrowprops=dict(arrowstyle="->", lw=1.5))

# Set the limits and hide the axes
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

plt.title("Proposed Methodological Flowchart", fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define key concepts
central_concept = "CVD Prediction"
concepts = [
    "AI Model",
    "Dynamic Simulation",
    "Streamlit App",
    "Ensemble Learning",
    "Clinical Validation",
    "User Trust",
    "Feature Engineering",
    "Bias Mitigation",
    "SMOTE"
]

# Define coordinates for the concepts
coordinates = {
    "CVD Prediction": (0.5, 0.5),
    "AI Model": (0.5, 0.8),
    "Dynamic Simulation": (0.7, 0.7),
    "Streamlit App": (0.3, 0.6),
    "Ensemble Learning": (0.7, 0.3),
    "Clinical Validation": (0.5, 0.2),
    "User Trust": (0.3, 0.4),
    "Feature Engineering": (0.6, 0.4),
    "Bias Mitigation": (0.2, 0.5),
    "SMOTE": (0.8, 0.5)
}

# Create the plot
fig, ax = plt.subplots(figsize=(10, 8))

# Plot the central concept
x, y = coordinates[central_concept]
ax.text(x, y, central_concept, fontsize=14, ha='center', va='center',
        bbox=dict(facecolor='orange', edgecolor='black', boxstyle='round,pad=0.5'))

# Plot each surrounding concept and draw connections
for concept, (x, y) in coordinates.items():
    if concept != central_concept:
        ax.text(x, y, concept, fontsize=12, ha='center', va='center',
                bbox=dict(facecolor='skyblue', edgecolor='black', boxstyle='round,pad=0.3'))
        ax.plot([coordinates[central_concept][0], x], [coordinates[central_concept][1], y], 'k--')

# Set limits and hide axes
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

plt.title("Graphical Abstract for AI-Driven CVD Prediction Methodology", fontsize=16)
plt.show()
```
